```python
pages=[i for i in range(301,652)]
content=[]
pages_divided=list(np.array_split(pages,4))
```

```python
def get_hrefs(pages_part):
    content=[]
    with r.Session() as f:
        for part in pages_part:
            raw_raw_raw_data=f.get(f'https://www.domofond.ru/prodazha-kvartiry-kazan-c1330?Page={part}',headers= ({'User-Agent': choice(desktop_agents)}),verify=False)
            soup=bs(raw_raw_raw_data.text,'lxml')
            raw_raw_data=soup.find_all('a',{'class':'long-item-card__item___ubItG'},href=True)
            for href in raw_raw_data:
                ref=re.search(r'href="(.*?)"',str(href))
                content.append(ref.group(1))
            #z=np.random.randint(4,10)
            print(f'Обработана страница {part} и {len(raw_raw_data)} квартиры.')
            #time.sleep(z)
    return content
```

```python
pool = ThreadPool(4)
l=pool.map(get_hrefs,pages_divided)
```

```python
references=[]
for part in l:
    for part_part in part:
        references.append(part_part)
```

```python
def get_info(references_part):
    dirty_html=[]
    k=0
    with r.Session() as s:
        for refer in references_part:
            code=s.get(f'https://www.domofond.ru'+refer,headers= ({'User-Agent': choice(desktop_agents)}),verify=False)
            soup=bs(code.text,'lxml')
            dirty_html.append(soup)
            k=k+1
            #z=np.random.randint(3,7)
            print(f'Обработал {k} предложение')
            #time.sleep(z)
    return dirty_html
```

```python
references_divided=list(np.array_split(references,4))
pool = ThreadPool(4)
dirty_htmls_list=pool.map(get_info,references_divided)
```

```python
codes=[]
for i in dirty_htmls_list:
    for j in i:
        codes.append(j)
```

```python
def get_info(codes):
    rf=[]
    d=[]
    p=[]
    m=[]
    da=[]
    #rooms square floor total_floors price_per_meter material,address
    for code in codes:
        try:
            rf.append(code.contents[1].find_all('h1',{'class':'information__title___1nM29'})[0].get_text())
        except:
            rf.append('0')
        try:
            d.append(code.contents[1].find_all('a',{'class':'area-rating__link___ZsnWR'})[0].get_text().replace('→','').lstrip().rstrip())
        except:
            try:
                d.append(code.contents[1].find_all('div',{'class':'area-rating__title___2n4JU'})[0].get_text().replace('→','').lstrip().rstrip())
            except:
                d.append('0')
        try:
            result1=re.findall(r'"details":\[{"name":"Тип","value".+?\]',str(code))[0]
            if "Цена за м²" not in result1:
                result1=re.findall(r'"details":\[{"name":"Тип","value".+?\]',str(code))[1]
            ppm=re.search(r'{"name":"Цена за м²","value":"(.*?)"',result1).group(1)
            mat=re.search(r'{"name":"Материал здания","value":"(.*?)"',result1).group(1)
            p.append(ppm)
            m.append(mat)
        except:
            p.append('0')
            m.append('0')
        try:
            da.append(code.contents[1].find_all('a',{'class':'information__address___1ZM6d'})[0].get_text())
        except:
            da.append('0')
    return (rf,d,p,m,da)
    
    
```

```python
rsftf,distr,price_per_meter,material,dirty_address=get_info(codes)
```

```python
rooms=[]
square=[]
floor=[]
total_floor=[]

for i in rsftf:
    if i!='0':
        temp=i.split(',')
        rooms.append(temp[0].lstrip().rstrip())
        square.append(temp[1].lstrip().rstrip())
        floor.append(temp[2].split('/')[0].lstrip().rstrip())
        total_floor.append(temp[2].split('/')[1].lstrip().rstrip().replace(' эт.',''))
    else:
        rooms.append('0')
        square.append('0')
        floor.append('0')
        total_floor.append('0')
```

```python
summary=pd.DataFrame([dirty_address,distr,material,floor,total_floor,square,rooms,price_per_meter]).T
summary.columns=['Адрес','Район','Материал','Этаж','Всего этажей','Площадь','Комнаты','Цена за кв.метр']
```

```python
summary.drop(summary[(summary['Адрес']=='0')|(summary['Материал']=='0')|(summary['Этаж']=='0')|(summary['Всего этажей']=='0')|(summary['Площадь']=='0')|(summary['Комнаты']=='0')].index,inplace=True)
```

```python
summary.index=[i for i in range(len(summary))]
```

```python
summary['Площадь']=summary['Площадь'].map(lambda x:x.replace('м²','').lstrip().rstrip())
```

```python
summary['Цена за кв.метр']=summary['Цена за кв.метр'].map(lambda x:x.replace('₽','').replace(' ','')).astype('int')
```
